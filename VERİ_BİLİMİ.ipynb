# AYKIRI DEĞER VAR MI TESPİT EDELİM

```python
# IQR hesaplama ve aykırı değer tespiti
# Sadece sayısal sütunları seçme
numeric_columns = df.select_dtypes(include=["float64", "int64"]).columns

for col in numeric_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR   ## ALT EŞİK
    upper_bound = Q3 + 1.5 * IQR   ## ÜST EŞİK

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"{col} değişkenindeki aykırı değer sayısı: {outliers.shape[0]}")
```

# **AYKIRI DEĞER PROBLEMİ ÇÖZÜMÜ**

```python
from sklearn.neighbors import LocalOutlierFactor
import pandas as pd

# Eksik verileri göz ardı ederek bağımsız değişkenleri seçme
X = df.dropna()  # Eksik veriler geçici olarak çıkarılır

# Sadece sayısal bağımsız değişkenleri seçme
X_numeric = X.select_dtypes(include=["float64", "int64"])

# LOF modelini tanımlama ve aykırı değerleri tespit etme
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)  # contamination=aykırı değer oranı
X_numeric["LOF_Score"] = lof.fit_predict(X_numeric)

# Aykırı değerleri (-1) olarak etiketleyen satırları belirleme
outliers = X_numeric[X_numeric["LOF_Score"] == -1].index   ## outliers değişkeni, aykırı değerlerin indekslerini bir pandas.Index nesnesi olarak döndürür.

# Aykırı değerlerin indekslerini liste halinde yazdırma
outlier_indices = list(outliers)
print(f"Aykırı Değer Sayısı: {len(outlier_indices)}")
print("Aykırı Değerlerin İndeksleri:\n", outlier_indices)

# Aykırı değerleri baskılama (medyan ile)
for col in X_numeric.columns[:-1]:  # LOF_Score hariç
    median_value = X_numeric.loc[X_numeric["LOF_Score"] != -1, col].median()  # Aykırı olmayanların medyanı
    X_numeric.loc[outliers, col] = median_value

print("\nAykırı değerler medyan ile baskılandı.")

# Orijinal veri setine temizlenmiş değerleri geri yazma
df.loc[X_numeric.index, X_numeric.columns[:-1]] = X_numeric[X_numeric.columns[:-1]]

# Temizlenmiş veri setini kaydetme
output_file = "temizlenmis_veri_seti.csv"
df.to_csv(output_file, index=False)

print(f"\nTemizlenmiş veri seti '{output_file}' dosyasına kaydedildi.")
print("\nVeri setindeki ilk birkaç satır:")
print(df.head())
```

# EKSİK VERİ ÇÖZÜMÜ : DEĞER ATAMA

```python
# Eksik veriler için sayısal ve kategorik değişkenlerin ayrılması
numeric_columns = df.select_dtypes(include=["float64", "int64"]).columns
categorical_columns = df.select_dtypes(include=["object"]).columns

# Sayısal değişkenlerde medyan ile eksik değer doldurma
for col in numeric_columns:
    median_value = df[col].median()
    df[col].fillna(median_value, inplace=True)

# Kategorik değişkenlerde mod (en sık değer) ile eksik değer doldurma
for col in categorical_columns:
    mode_value = df[col].mode()[0]
    df[col].fillna(mode_value, inplace=True)

# Eksik veriler çözüldükten sonra kontrol
print("Eksik Veri Kalmadı mı?:\n", df.isnull().sum().sum())



# Temizlenmiş veri setini kaydetme
df.to_csv("eksik_veri_cozulmus_veri_seti.csv", index=False)
print("Eksik veri problemi çözülmüş veri seti kaydedildi.")
```

# **SINIFLANDIRMA **

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler

# Modelleri içe aktaralım
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

# RANDOM_STATE belirli bir sabit değerle kullanılırsa, her çalıştırmada aynı rastgelelik sonuçları elde edilir.
RANDOM_STATE = 42

# Veri setini yükleme
# "donusum_tamamlanmis_veri_seti" veri setini yükleyelim (örnek olarak bir CSV dosyasından):
df = pd.read_csv("donusum_tamamlanmis_veri_seti.csv")

# Bağımsız (X) ve bağımlı (y) değişkenleri ayıralım
# "target" sütunu hedef değişken olarak varsayılmıştır, değiştirmeniz gerekebilir
X = df.drop(columns=["Workout_Type"])
y = df["Workout_Type"]

# Eğitim ve test setlerine ayıralım
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=RANDOM_STATE)

# Veriyi ölçeklendirme (KNN ve SVM için önemli)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Modeller ve hiperparametreler
models = {
    "Logistic Regression": {
        "model": LogisticRegression(random_state=RANDOM_STATE),
        "params": {"C": [0.1, 1, 10, 100], "solver": ["liblinear", "lbfgs"]}
    },
    "K-Nearest Neighbors": {
        "model": KNeighborsClassifier(),
        "params": {"n_neighbors": [3, 5, 7, 9], "weights": ["uniform", "distance"]}
    },
    "Support Vector Machine": {
        "model": SVC(random_state=RANDOM_STATE),
        "params": {"C": [0.1, 1, 10], "kernel": ["linear", "rbf"]}
    },
    "Decision Tree": {
        "model": DecisionTreeClassifier(random_state=RANDOM_STATE),
        "params": {"max_depth": [None, 5, 10], "criterion": ["gini", "entropy"]}
    }
}

# Modelleri eğitip test etme
best_models = {}
for model_name, mp in models.items():
    print(f"\nModel: {model_name}")
    grid = GridSearchCV(mp["model"], mp["params"], cv=5, scoring='accuracy')
    grid.fit(X_train, y_train)

    # En iyi model ve hiperparametreler
    print(f"En iyi hiperparametreler: {grid.best_params_}")
    best_model = grid.best_estimator_
    best_models[model_name] = best_model

    # Test setinde modelin performansı
    y_pred = best_model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"Test Seti Doğruluk: {acc:.2f}")
    print("Sınıflandırma Raporu:")
    print(classification_report(y_test, y_pred))

# Tüm modellerin karşılaştırılması
print("\nModel Performans Özeti:")
for name, model in best_models.items():
    acc = accuracy_score(y_test, model.predict(X_test))
    print(f"{name}: {acc:.2f}")
```

```python
from sklearn.metrics import precision_score, f1_score, recall_score

# Performans ölçütlerini hesaplamak için bir sözlük
performance_summary = {
    "Model": [],
    "Test Accuracy": [],
    "Ortalama Precision": [],
    "Ortalama F1-Score": [],
    "Her Sınıf için Recall": []
}

# Her bir modelin performansını hesapla
for model_name, model in best_models.items():
    # Tahminler
    y_pred = model.predict(X_test)

    # Performans ölçütlerini hesaplama
    test_acc = accuracy_score(y_test, y_pred)
    avg_precision = precision_score(y_test, y_pred, average='macro')  # Ortalama precision
    avg_f1 = f1_score(y_test, y_pred, average='macro')  # Ortalama F1-Score
    class_recall = recall_score(y_test, y_pred, average=None)  # Her sınıf için recall

    # Sonuçları saklama
    performance_summary["Model"].append(model_name)
    performance_summary["Test Accuracy"].append(test_acc)
    performance_summary["Ortalama Precision"].append(avg_precision)
    performance_summary["Ortalama F1-Score"].append(avg_f1)
    performance_summary["Her Sınıf için Recall"].append(class_recall)

# Performans özetini bir tablo olarak gösterelim
performance_df = pd.DataFrame(performance_summary)
print("\nModel Performans Özeti:")
print(performance_df)

# Her model için sınıf bazında recall'ları yazdır
for i, row in performance_df.iterrows():
    print(f"\nModel: {row['Model']}")
    for idx, val in enumerate(row["Her Sınıf için Recall"]):
        print(f"  Sınıf {idx} Recall: {val:.2f}")
```

# **EN İYİ PERFORMANS DEĞERİNE SAHİP MODELİN BOYUTUNU AZALTMA**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, f1_score, recall_score
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Modelleri içe aktaralım
from sklearn.linear_model import LogisticRegression

# Rastgelelik sabitleyelim RANDOM_STATE belirli bir sabit değerle kullanılırsa, her çalıştırmada aynı rastgelelik sonuçları elde edilir.
RANDOM_STATE = 42

# Veri setini yükleme
# "donusum_tamamlanmis_veri_seti" veri setini yükleyelim (örnek olarak bir CSV dosyasından):
df = pd.read_csv("donusum_tamamlanmis_veri_seti.csv")

# Bağımsız (X) ve bağımlı (y) değişkenleri ayıralım
# "Workout_Type" sütunu hedef değişken olarak kullanılıyor
X = df.drop(columns=["Workout_Type"])
y = df["Workout_Type"]

# Eğitim ve test setlerine ayıralım
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=RANDOM_STATE)

# Veriyi ölçeklendirme (KNN ve SVM için önemli)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Performansı en yüksek modeli manuel olarak seçiyoruz
# Burada örnek olarak Logistic Regression modelini seçiyoruz
best_model_name = "Logistic Regression"
best_model = LogisticRegression(random_state=RANDOM_STATE)

# Modeli orijinal veri setiyle eğitip test edelim
best_model.fit(X_train, y_train)
y_pred_original = best_model.predict(X_test)

# Orijinal veri setindeki performans ölçümleri
original_accuracy = accuracy_score(y_test, y_pred_original)
original_precision = precision_score(y_test, y_pred_original, average='macro')
original_f1 = f1_score(y_test, y_pred_original, average='macro')
original_recall = recall_score(y_test, y_pred_original, average='macro')

# Boyut azaltma (PCA) uygulayalım.Temel amacı, verideki bilgi kaybını en aza indirerek veri boyutunu küçültmektir.
pca = PCA(n_components=2)  # Boyutu 2 bileşene indirelim
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

# Modeli boyut azaltılmış veri setiyle eğitip test edelim
best_model_pca = LogisticRegression(random_state=RANDOM_STATE)  # Aynı model ile PCA sonrası eğitim
best_model_pca.fit(X_train_pca, y_train)
y_pred_pca = best_model_pca.predict(X_test_pca)

# Boyut azaltılmış veri setindeki performans ölçümleri
pca_accuracy = accuracy_score(y_test, y_pred_pca)
pca_precision = precision_score(y_test, y_pred_pca, average='macro')
pca_f1 = f1_score(y_test, y_pred_pca, average='macro')
pca_recall = recall_score(y_test, y_pred_pca, average='macro')

# Performans karşılaştırması

print("\nPerformans Karşılaştırması:")
print(f"Orijinal Veri Seti - Accuracy: {original_accuracy:.2f}, Precision: {original_precision:.2f}, F1-Score: {original_f1:.2f}, Recall: {original_recall:.2f}")
print(f"PCA ile Azaltılmış Veri Seti - Accuracy: {pca_accuracy:.2f}, Precision: {pca_precision:.2f}, F1-Score: {pca_f1:.2f}, Recall: {pca_recall:.2f}")
```
